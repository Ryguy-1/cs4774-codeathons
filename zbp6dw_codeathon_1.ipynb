```python
# --- (4) Does it contain any categorical data? ---
categorical_features_names = housing_df.select_dtypes(include=['object']).columns
print(f"Number of categorical features: {len(categorical_features_names)}")
print(f"All categorical features: {categorical_features_names}")
print(f"Contains categorical data: {'Yes' if len(categorical_features_names) > 0 else 'No'}")

# for each categorical feature, print all possible values
for feature in categorical_features_names:
    print(f"Feature: {feature}")
    print(housing_df[feature].value_counts())
    print()
```
